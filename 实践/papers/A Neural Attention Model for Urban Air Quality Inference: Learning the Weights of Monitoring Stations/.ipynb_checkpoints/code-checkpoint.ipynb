```python
## get distance between any two stations
dist = np.zeros((len(station),len(station)))
angle = np.zeros((len(station),len(station)))
tmp_dist = []
tmp_angle = []
for i, loc1 in enumerate(zip(station['latitude'],station['longitude'])):
    for j, loc2 in enumerate(zip(station['latitude'],station['longitude'])):
        if i != j:
            tmp_dist.append(getDistance(loc1[1],loc1[0],loc2[1],loc2[0]))
            tmp_angle.append(getAngle(loc1[0],loc1[1],loc2[0],loc2[1]))
            tmp_angle.append(getAngle(loc2[0],loc2[1],loc1[0],loc1[1]))
```

```python
### do scaler
tmp_dist = np.array(tmp_dist)
tmp_angle = np.array(tmp_angle)
dist_scaler = MinMaxScaler(min = np.min(tmp_dist),max = np.max(tmp_dist))
angle_scaler = MinMaxScaler(min = np.min(tmp_angle),max = np.max(tmp_angle))
tmp_dist = dist_scaler.transform(tmp_dist)
tmp_angle = angle_scaler.transform(tmp_angle)
```

```python
tmp_dist
```

```python
## copy tmp_dist/tmp_angle to dist/angle
index = 0
for i in range(len(station)):
    for j in range(len(station)):
        if i != j:
            dist[i][j] = tmp_dist[index]
            angle[i][j] = tmp_angle[index]
            index += 1
```

```python
pd.DataFrame(dist).to_csv('dist.csv')
pd.DataFrame(angle).to_csv('angle.csv')
```

```python
### build index for station_id
station2district = dict()
for station_id, district_id in zip(station['station_id'],station['district_id']):
    station2district[station_id] = district_id
```

```python
station2district
```

```python
date = pd.date_range('20140501','20150501',freq = 'H')
date = np.array(date.format())
date = date[:-1]
```

```python
for key,value in station2district.items():
    save_frame = mete[mete['id'] == value]
    time = np.array(save_frame['time'])
    date_index = 0;time_index = 0
    while time_index < len(time) and date_index < len(date):
        if time[time_index] != date[date_index]:
            tmp_frame = pd.DataFrame([date[date_index]],columns = ['time'])
            save_frame = pd.concat([save_frame,tmp_frame],ignore_index = True,axis = 0)
            date_index += 1
        else:
            date_index += 1
            time_index += 1
#         print('hello world')
    print(save_frame.shape, '--- ',key)
    save_frame.fillna(0,inplace = True)
    save_frame = save_frame.sort_values(by = 'time')
    save_frame = save_frame.iloc[:-1]
    save_frame.to_csv('mete_' + str(key) + '_norm.csv',index = False,encoding = 'gb2312')
```
