```python
plt.figure(figsize =  (4,4))
labels = ['no','yes']
sizes = arr / np.sum(arr)
patches, l_text, p_text = plt.pie(sizes, labels = labels,
                                  labeldistance = 1.05, autopct = '%3.1f%%', shadow = False,
                                  startangle = 90, pctdistance = 0.6)
for t in l_text:
    t.set_size = 30
for t in p_text:
    t.set_size = 20
plt.axis('equal')
plt.legend()
# plt.show()
plt.savefig('dis.jpg')
```
